```python
from antlr4 import *
from Python3Lexer import Python3Lexer
from Python3Parser import Python3Parser
from Python3ParserListener import Python3ParserListener


class TreePrinter(Python3ParserListener):
    def __init__(self):
        self.indent = 0

    def enterEveryRule(self, ctx):
        rule_name = Python3Parser.ruleNames[ctx.getRuleIndex()]
        print('  ' * self.indent + rule_name)
        self.indent += 1

    def exitEveryRule(self, ctx):
        self.indent -= 1

    def visitTerminal(self, node):

        text = node.getText()
        if text not in ["\n", "\r\n", "\t", " "]:
            print('  ' * self.indent + f'TOKEN({node.symbol.type}): {text}')

def main():
    code = """
def hello():
    print("Hello, world!")

x = 2+3
y = 4+6

print(x*y)

class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("John", 36)

print(p1.name)
print(p1.age)
"""
    input_stream = InputStream(code)
    lexer = Python3Lexer(input_stream)
    stream = CommonTokenStream(lexer)
    parser = Python3Parser(stream)
    tree = parser.file_input()
    printer = TreePrinter()
    walker = ParseTreeWalker()
    walker.walk(printer, tree)



main()
```
